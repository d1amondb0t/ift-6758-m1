```python
import pprint
import load_game
import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from pathlib import Path
from IPython.display import display
```

```python
# --- your existing setup ---
# widget 1 for game ID ✅
# widget 2 for switch between regular & playoffs ✅
# widget 3
#first 4 digits = year" ✅
#next 2 digits = 01 (preseason), 02 = (regular season), 03 = playoffs, 04 = all-star" ✅
#final 4 digits = game number ranging from 0001 to 2000" ✅
root = "dataStore"
ring = "../../figures/nhl_rink.png"

year = "2016"
year = widgets.IntSlider(
  min=2016,
  max=2023,
  value=2016,
  description='year'
)

game_id = widgets.IntSlider(
  min=1,
  max=2000,
  value=1,
  description='game_id'
)

season = widgets.ToggleButtons(options=[('Regular', 2), ('Playoffs', 3)], value=2, description='Season')
data = load_game.load_json(root, year.value, season.value, game_id.value)

plays = load_game.load_plays(data)
min_event_id, max_event_id = load_game.event_id_min_max(plays=plays)

event_id = widgets.IntSlider(
  min=min_event_id,
  max=max_event_id,
  value=min_event_id,
  description="Play #"
)

out = widgets.Output()
display(game_id, year, season, event_id, out)
observer = load_game.on_game_change(root, ring, year, season, game_id, event_id, out)

year.observe(observer, names="value")
game_id.observe(observer, names="value")
season.observe(observer, names="value")
event_id.observe(observer, names="value")

# initial draw
observer(None)
```
